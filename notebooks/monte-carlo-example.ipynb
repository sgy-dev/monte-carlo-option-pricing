# Monte Carlo Methods

A visual guide to Monte Carlo methods for use in pricing options contracts.

```python
from montecarlo.simulate import EuropeanOption, MonteCarloSimulator
```

```python
euro_option = EuropeanOption(
    S0 = 100.00,  # Initial stock price
    K = 105.00,   # Strike price
    T = 1.0,      # Time to maturity in years
    r = 0.05,     # Risk-free interest rate
    sigma = 0.2   # Volatility of the underlying asset
)
```

```python
mc = MonteCarloSimulator(
    option = euro_option,
    num_t_steps = 250,
    num_paths = 500
)
```

```python
mc.run_simulations()
mc.plot_simulations()
```
